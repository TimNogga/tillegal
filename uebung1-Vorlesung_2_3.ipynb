Logische Datenunabhängigkeit in der Drei-Schichten-Architektur ermöglicht es, das Datenbankdesign zu ändern, ohne Anwendungen anzupassen. UPS ist stolz darauf immer aktuelle Informationen über die Verarbeitung und den aktuellen Standort jedes ihrer versendeten Artikel zu haben. Dafür setzt UPS ein unternehmensweites Informationssystem ein. Versendete Artikel sind das Herzstück des UPS Produktverfolgungs-Informationssystems. Versendete Artikel können durch ihre Artikelnummern(eindeutig), Gewichte, Maße, Versicherungssumme, Zielorte und entgültige Lieferdaten bestimmt werden.
Versendete Artikel werden in einem einzigen Einzelhandelszentrum in das UPS-System aufgenommen. Einzelhandelszentren zeichnen sich durch ihren Typ, EindeutigeID und Adresse aus. Versendete Artikel gelangen über ein oder mehrere Standard-UPS-Transportereignisse (z.B. Flüge, LKW-Lieferungen) an ihren Bestimmungsort. Diese Transportereignisse sind durch eine eindeutige Zeitplannummer, einen Typ (z.B. Flug, LKW) und eine Lieferroute gekennzeichnet.
