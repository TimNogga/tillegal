```python
import numpy as np
```

```python

def least_squares(x,y):
    A = np.zeros((10,10))
    for i in range(10):
        for j in range(10):
            A[i,j] = x[i]**j
    b = y
    solution = np.linalg.solve(A.T@A,A.T@b)
    return A,solution
def regularizedMatrixT(x,y,lam):
    A = np.zeros((10,10))
    for i in range(10):
        for j in range(10):
            A[i,j] = x[i]**j
    b = y
    temp = np.zeros((20,10))
    temp[0:10,:] = A
    temp[10:20,:] = lam*np.eye(10)
    solution_2 = np.linalg.solve(temp.T@temp,A.T@y)
    return solution_2
```
